## __What does R do well__

Apart from having a large number of libraries which are designed to implement almost any statistical or modelling techniques you are likely to want to, there are some things R is optimised to do well which will affect the way you design your code.

In particular R is an interpreted language in which functions can be written in a compiled language - what does that mean?

An interpreted language reads your code and converts it to instructions that the computer will understand step by step through your code. This process of reading and translating is slow but has the advantage that the code can be easily changed and you can work with R like you would with a calculator (like we are in these notebooks).

A compiled language does the same thing but all the conversion to computer instructions is done before you run the program. This can take a long time to do and if you want to change __anything__ you have to compile the entire code again. However, when you run the code, it runs as fast as is possible and that can orders of magnitude faster - sometimes hundreds of times faster.

So, if you want to do something in R that requires multiple operations, say calculate the square-root of each value in a vector, in most languages you would do this by using a function on each item in the vector. In R this is really slow, but the functions have been specially written to work on all the entries in a vector on one function call and that is really fast.

The following is a bit of a contrived example, but it illustrates the point. Don't worry about the details of the code, we will cover all of this in later notebooks.

```r
n <- 10000000
x <- seq(1:n)
y <- rep(0, n)
# Option 1: calculate each entry one at a time.
start <- as.numeric(Sys.time())
for (j in x) { y[j] <- sqrt(x[j]) ^ 2}
end <- as.numeric(Sys.time())
cat("Individual entries took", end-start, "s.\n")
# Option 2: calculate all entries in one call.
start <- as.numeric(Sys.time())
z <- sqrt(x) ^ 2
end <- as.numeric(Sys.time())
cat("Vector calculation took", end-start, "s.\n")
```

What this means in practice is that you should try to use R to perform the same operation on all rows in a data frame rather than to work through the data frame one row at a time. There are two kinds of operations to help with this: the base-R apply functions to get data from simple tables and the tidyverse's table manipulation (wrangling) functions.

## __apply()__

If you need to do something to many rows of data at the same time, you can use R's _apply_ functions: __apply()__, __sapply()__ and __lapply()__.

The apply() function allows you to perform the same operation on each row, column or both in a data frame. You specify which using the MARGIN parameter: MARGIN=1 applies to rows, MARGIN=2 applies to columns and MARGIN=c(1, 2) applies to each cell. For example, if we have a sequence of test result for a set of participants over a period of time, we can get the mean() values for each participant, at each sampling point or round the data in each cell.

Note that in the last case, we are calling the round() function which takes a value AND the number of decimal places to round to. The first parameter used is the cell id, all the remaining parameters need to be supplied to the apply call, which is why 'round' is followed by '2' - trying changing the '2' to a different value, for example, '0'.

```r
n <- 10
patient_1 <- rnorm(n, mean=10, sd=1)
patient_2 <- rnorm(n, mean=12, sd=2)
patient_3 <- rnorm(n, mean=2, sd=1)
patient_4 <- rnorm(n, mean=5, sd=5)
patient_5 <- rnorm(n, mean=8, sd=3)
# t() converts a column vector to a row (transpose - see matrices).
tests <- data.frame(id=1, t(patient_1))
tests[2, ] <- data.frame(id=2, t(patient_2))
tests[3, ] <- data.frame(id=3, t(patient_3))
tests[4, ] <- data.frame(id=4, t(patient_4))
tests[5, ] <- data.frame(id=5, t(patient_5))
tests
```

So now, we can run the apply() function to get row, column or both results.

```r
# Row means.
apply(tests[, 1:n+1], MARGIN=1, mean)
```

```r
# Column sums.
apply(tests[, 1:n+1], MARGIN=2, sum)
```

```r
# Cell rounding using the round() function which takes two parameters
# the thing with the data in and the number of decimal places to round
# to e.g. round(number, 2) will return the value of number rounded to
# two decimal places (if number is 1.2345, it will return 1.23).
apply(tests[, 1:n+1], MARGIN=c(1, 2), round, 0)
```

The other apply() functions are simply variations in the way the output from the function is created. lapply() produces a list output (and usually has a list input but doesn't have to - although, as you will see below, it does just apply to the columns) and sapply produces a similar vector output.

```r
# Column sums.
lapply(tests[, 1:n+1], mean)
```

```r
# Column sums.
sapply(tests[, 1:n+1], mean)
```

## __Tidyverse__

The tidyverse is a very useful and well documented library designed for data science. See [R for Data Science](https://r4ds.had.co.nz). The library includes a large number of functions that allow you to process a data frame efficiently and in a very readable way.

First of all, you need to load the library (if it is not installed, type, `install.packages("tidyverse")`).

```r
library(tidyverse)
```

To demonstrate the tidyverse, we'll use a useful dataset taken from those included in R - to find out what is there, type `data()`. We'll use the `starwars` dataset (to see the first few rows of a data frame, you can use `head()`).

You may also notice that the data frames are often referred to as `tibble`s. This is a special tidyverse version of a data frame with a little extra functionality - for example, it doesn't display too much data. You can convert from tibbles to data frames using as_tibble() and as.data.frame() if you really need to.

```r
as_tibble(as.data.frame(head(starwars)))
```

### _Tidy data_

Before we continues, it is worth mentioning a key requirement of good primary data tables/data frames which is intended to make life simple, effective and reliable when it comes to data science:

Each row will contain data for just one record - i.e. one patient, one participant etc. Each column will contain data for just one variable - e.g. age, gender, a test result etc. And, finally, each cell in a table will contain just one piece of data.

Sometimes, files do not match this requirement for other design reasons, but it is always worth bearing this in mind as a goal certainly for the tables you spend most time working with and your platform file.

If you look at the `starwars` data frame, there is one and only one row for each character, there is one and only one column for each feature (or variable). Mostly there is only one item in each cell (the lists being the exception). Such lists are a classic problem - we may need to make room for 10 diagnosis codes in complex cases, but if most people only have two, that's a waste of memory, so there are often trade-offs that we need to handle.

With a data frame organised in the one row per record, one columne per variable, one value per cell way, the following processes become very efficient.

### _Selecting features of a data frame_

There are two kinds of selection we do regularly with data frames: select rows/records of interest and columns/variables of interest. The tidyverse provides two function to do this: __filter()__ to select rows and __select()__ to select columns.

We filter to select only those records that we are interested in, for example, we can select all the human characters.

```r
humans <- filter(starwars, species=='Human')
head(humans)
```

We select columns to simplify the table (often as a precursor to some serious processing of tables that aren't in the tidy format to extract data which is more compatible to a tidy format - for example, finding the range of test results rather than listing them all).

```r
small_starwars <- select(starwars, c(name, species, homeworld))
head(small_starwars)
```

And, of course, we can do both things, on after another - in this case, we select the columns first and then, using the resulting data frame as the parameter for the filter function, select only the humans from that data frame:

```r
humans_smallstarwars <- filter(select(starwars, c(name, species, homeworld)), species=='Human')
head(humans_smallstarwars)
```

#### _The pipe, `%>%`_

That last line is a bit difficult to read and when you combine it (as you often will) with the other tidyverse functions, expressions can quickly become almost incomprehensible - you can see that the filter command at the far left of the instruction is associated with the species parameter on the far right.

The tidyverse has a very neat way around this - it is the 'pipe' operator, `%>%`. You can think of the pipe as taking the output data frame from whatever is to its left and feeding that data frame to the operation that is to its right. The result is that you can read the code from left to right very simply.

```r
humans_smallstarwars <- starwars %>% select(c(name, species, homeworld)) %>% filter(species=='Human')
head(humans_smallstarwars)
```

And, as with the base-R code, we can then apply functions to the selected data, for example, we can calculate the mean of numeric columns. Note that the parameter na.rm which can be defined for the mean() function instructs mean() to ignore any NA (not available) values it finds (otherwise, if R makes any calculation with an NA, the result is NA which is not usually useful).

```r
starwars %>%
    select(c(name, species, homeworld, height, mass)) %>%
    filter(species=='Human') %>%
    summarise(n=n(), mean_h=mean(height, na.rm=TRUE), mean_m=mean(mass, na.rm=TRUE))
```

We'll use quite a lot of the tidverse functions over the course, the most useful ones being:

* arrange() - sorts the data frame by the specified columns.
* mutate() - creates a new column/variable.
* group_by() - groups the data so that analyses apply to the groups rather than the whole table.

#### _arrange()_

The arrange() function sorts the data frame in order of the specified columns in the order specified such that the first column is used to sort the data, then the next is used to sort the data within those categories, and then the third within those two categories and so on.

For example, sorting by species, homeworld and height (in reverse order), we get:

```r
starwars %>%
arrange(species, homeworld, -height)
```

#### _mutate()_

mutate() is used to modify data in a column (by using an existing column name) or to create a new column of data (by using a new column name). For example, we could create a new variable indicating whether the character has hair or not. First of all we'll replace all the NAs with 'none' (a column update), then we'll look for 'none' to indicate no hair (a new column).

To select a particular value, we use the ifelse() function which is similar to the if ... else code we will cover in the fifth notebook. It works out the expression in the first parameter, here

is.na(hair_color)   - which means is the entry in hair_color equal to `NA` (there's a technical reason we can't just say hair_color == NA which we'll go into during the module).

If this works out to be TRUE as it is for R2-D2, then this rows entry in the specified column is set (chnaged) to the value of the next parameter, here `none`. If it is FALSE as it is for Luke Skywalker, then the entry in the column is set to the value of hair_color (i.e. left unchanged, as it is).

The second ifelse() causes the entry for each row to be set to 0 if that row has hair_color set to 'none' and 1 otherwise.

If this is not making sense, don't panic, review this section again after you have worked through notebook 5 and see if that helps.

```r
starwars_hair <- starwars %>% mutate(hair_color=ifelse(is.na(hair_color), 'none', hair_color))
starwars_hair <- starwars_hair %>% mutate(has_hair=ifelse(hair_color=='none', 0, 1))
head(starwars_hair)
```

#### _group_by()_

This function is often the hardest to get to grips with. It configures a data frame (and it is not always clear that it has been configured) so that it is grouped by the columns specified. This is like the arrange() function which just orders the rows, but group_by() causes any future summary functions to be applied to the specified groups rather than the whole table. For example, after a group_by() the mean value of the height would be the mean value for each group. You can remove this grouping using the ungroup() function - I do this by default every time I use group_by() so that it is more difficult to accidently leave a data framed in a grouped state.

The following code gets the mean height for the whole table:

```r
starwars %>% summarise(mean_h=mean(height, na.rm=TRUE))
```

But the following code gets the mean height for each species:

```r
starwars %>% group_by(species) %>% summarise(mean_h=mean(height, na.rm=TRUE)) %>% ungroup()
```
